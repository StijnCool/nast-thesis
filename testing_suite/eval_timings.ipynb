```python
f=open("timings_5gen_10pop.txt",'r')
s = f.read()
f.close()
s = s.split('\n')
s = np.array([ss[20:-1].split() for ss in s]).astype(float).transpose()
s[0] = s[0]/5
s = np.append(s, [s[0]-np.sum(s[1:], axis=0)], axis=0)
print(s)
titles = ['total','selection','mutation','shuffle','crossover','fitness','total - rest (save, counter, loop)']

for i in range(0,s.shape[0]):
    ss = s[i]
    weights = np.ones_like(ss) / ss.size
    plt.figure()
    plt.hist(ss, weights=weights, rwidth=.9)
    plt.title(titles[i])
    plt.xlabel('time (s)')
    plt.show()
    print(np.mean(ss))
```
